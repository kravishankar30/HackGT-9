```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
na_vals = ['NA' ,'Missing', 'N.A', 'N.A.', 'Null', 'null', 'missing', 'None']
def csvgrapher(csvName, plot, xaxis, yaxis):
    df = pd.read_csv(csvName, na_values = na_vals)
    df.dropna(subset = [xaxis, yaxis], inplace = True, how = 'any')
    if plot == 'scatter':
        plt.scatter(df[xaxis], df[yaxis], edgecolor = 'black', linewidth = 0.5, alpha = 0.75)
        
        plt.xlabel(xaxis)
        plt.ylabel(yaxis)
        plt.tight_layout()
        plt.show()
    elif plot == 'line plot':
        plt.plot(df[xaxis], df[yaxis], )
        plt.xlabel(xaxis)
        plt.ylabel(yaxis)
```

```python
csvgrapher('BR_youtube_trending_data.csv', 'scatter', 'view_count', 'likes')
```
